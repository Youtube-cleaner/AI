```python
import pandas as pd

data = pd.read_csv('LSTM_Dataset.csv')
```

```python
data.columns = ['document']
```

```python
data
```

```python
label_list = []
for l in range(len(data)):
  for i in range(len(data['document'][l])):
  
    if(data['document'][l][i] == '\t'):
      label_list.append(data['document'][l][i+1])
      
    
```

```python
label_list
```

```python
data['label'] = label_list
```

```python
import re
```

```python
def data_preprocessing(sentence):
  sentence_split = sentence.split()

  sentence_split_result = []

  for word in sentence_split:
   if not(word[-1].isdigit()):
     sentence_split_result.append(word)

  sentence_result = " ".join(sentence_split_result)


  return (sentence_result)

```

```python
data["document"] = data["document"].apply(data_preprocessing)       #전체 데이터에 적용
```

```python
data
```

```python
data.loc[data.label == '1', 'label'] = 'non hate'
```

```python
data.loc[data.label == '0', 'label'] = 1
```

```python
data.loc[data.label == 'non hate', 'label'] = 0
```
