# **DataFrame으로 열기**

```python
import pandas as pd
```

```python
data = pd.read_csv('hate_speech_binary_dataset.csv')
```

```python
data.columns = ['document','value']
```

# **결측값 확인, 삭제**

```python
print(data.isnull().any())
```

```python
data = data.dropna()
```

```python
data[(data.document).isna()]
```

# **Value 값 반전**

```python
data
```

```python
data[(data.value) == 1] 
```

```python
data.loc[data.value == 1, 'value'] = 'non hate'
```

```python
data.loc[data.value == 0, 'value'] = 1
```

```python
data.loc[data.value == 'non hate', 'value'] = 0
```

```python
data
```

```python
def word_split(x):
  return len(x.split())

data['word_count'] = data['document'].apply(word_split)
```
